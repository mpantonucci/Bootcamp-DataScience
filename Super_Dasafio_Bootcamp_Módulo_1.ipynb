```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Impacto da Pandemia de COVID no diagnóstico e tratamento do Cancêr de Mama (Brasil 2021)


Casos por Ano do tratamento segundo UF do tratamento

Diagnóstico Detalhado: C50 - Neoplasia maligna da mama , D05 - Carcinoma in situ da mama

Ano do diagnóstico: 2013-2021



```python
#leitura e adaptação do arquivo que será utilizado para a análise
#O Arquivo csv, contém a quantidade de casos em tratamento para o Cancer de Mama, por Estado, no Brasil, nos anos de 2013 até 2021.
dados_tratamento_arquivo_brasil = pd.read_csv("/content/drive/MyDrive/Cursos/Bootcamp Data Science/casos_cancer_mama_brasil.csv", sep=';', encoding="ISO-8859-1")

#dropar as colunas 'Total' e 'Sem informação de Tratamento' por que não existem informações que ajudem na análise sobre os casos de Cancer de Mama / 2021
dados_tratamento_arquivo_brasil = dados_tratamento_arquivo_brasil.drop(' Total', axis=1)
dados_tratamento_arquivo_brasil = dados_tratamento_arquivo_brasil.drop('Sem informação de tratamento', axis=1)

#definir o index do dataframe
dados_tratamento_arquivo_brasil = dados_tratamento_arquivo_brasil.set_index('UF do tratamento')

#dropar a linha Total 'Total' e 'Sem informação de Tratamento' por que não existem informações que ajudem na análise sobre os casos de Cancer de Mama / 2021
dados_tratamento_arquivo_brasil = dados_tratamento_arquivo_brasil.drop('Total', axis=0)
dados_tratamento_arquivo_brasil = dados_tratamento_arquivo_brasil.drop('Sem informação de tratamento', axis=0)
```

**Gráfico que exibe a quantidade de Tratamentos, para o Cancêr de Mama, realizados no estado de SP, no período de 2013 até 2021.**
> É possível, claramente, notar a diminuição de casos para o ano de 2021. 
A curva começa em declínio no ano de 2020, ano de início da Pandemia. 
O fato da curva está em declínio devido à pandemia é uma suposição, outros fatores desconhecidos podem ter interferido.

```python
#obter os casos de tratamento de cancer de Mama para o Estado de SP
dados_tratamento_arquivo_sp = dados_tratamento_arquivo_brasil[dados_arquivo_brasil.index.str.contains('São Paulo')]

ax = dados_tratamento_arquivo_sp.T.plot(figsize=(10,5))
ax.legend(loc="upper right", bbox_to_anchor=(1.20, 1))
plt.style.use("ggplot")
plt.title("Casos de Tratamento para o Cancer de Mama - Estado SP ")
plt.ylabel("Quantidade de Tratamentos")
plt.xlabel("Ano")
```

**Gráfico que exibe a quantidade de Tratamentos, para o Cancêr de Mama, para a região Sudeste no período de 2013 até 2021.**
> É possível, claramente, notar a diminuição de casos para o ano de 2021. A curva começa em declínio no ano de 2020, ano de início da Pandemia.
Nota-se que o declínico de casos em Tratamento ocorre para todos os estados.
O fato da curva está em declínio devido à pandemia é uma suposição, outros fatores desconhecidos podem ter interferido.

```python
#comparar o estado de SP com a região SUDESTE
dados_tratamento_regiao_sudeste = dados_tratamento_arquivo_brasil[dados_tratamento_arquivo_brasil.index.str.startswith('3')]

ax = dados_tratamento_regiao_sudeste.T.plot(figsize=(10,5))
ax.legend(loc="upper right", bbox_to_anchor=(1.23, 1))
plt.style.use("ggplot")
plt.title("Casos de Tratamento para o Cancer de Mama - Região Sudeste ")
plt.ylabel("Quantidade de Tratamentos")
plt.xlabel("Ano")
```

**Gráfico que exibe a quantidade de Tratamentos, para o Cancêr de Mama, para todo o Brasil em comparativo com o estado de SP no período de 2013 até 2021.**

> É possível, claramente, notar a diminuição de casos para o ano de 2021. A curva começa em declínio no ano de 2020, ano de início da Pandemia.Nota-se que o declínico de casos em Tratamento ocorre para o todo o Brasil.
O fato da curva está em declínio devido à pandemia é uma suposição, outros fatores desconhecidos podem ter interferido.

```python
#Comparar o estado de SP com o Total Brasil
dados_tratamento_total_brasil = pd.DataFrame(dados_tratamento_arquivo_brasil.drop('35 São Paulo', axis=0).sum(), columns=["Total"])
dados_tratamento_total_brasil = dados_tratamento_total_brasil.T
dados_tratamento_total_brasil = pd.merge(dados_tratamento_total_brasil, dados_tratamento_arquivo_sp, how='outer')
dados_tratamento_total_brasil = dados_tratamento_total_brasil.T
dados_tratamento_total_brasil = dados_tratamento_total_brasil.rename(columns={0: 'Brasil', 1: 'SP'})
ax = dados_tratamento_total_brasil.plot(figsize=(10,5))
ax.legend(loc="upper right", bbox_to_anchor=(1.08, 1))
plt.style.use("ggplot")
plt.title("Casos de Tratamento para o Cancer de Mama - Comparativo Estado SP x Brasil ")
plt.ylabel("Quantidade de Tratamentos")
plt.xlabel("Ano")

```

# Análise de Total de Pacientes diagnosticados com Cancêr de Mama 
Casos por Ano do diagnóstico segundo UF do tratamento

Diagnóstico Detalhado: C50 - Neoplasia maligna da mama , D05 - Carcinoma in situ da mama

Ano do diagnóstico: 2013-2021

```python
#leitura e adaptação do arquivo que será utilizado para a análise
#O Arquivo csv, contém a quantidade de casos de tratamento para o Cancer de Mama, por Estado, no Brasil, nos anos de 2013 até 2021.
diagnosticos_cancer_brasil = pd.read_csv("/content/drive/MyDrive/Cursos/Bootcamp Data Science/diagnosticos_cancer_mama_brasil.csv", sep=';', encoding="ISO-8859-1")

#dropar as colunas 'Total' por que não existem informações que ajudem na análise sobre os casos de Cancer de Mama / 2021
diagnosticos_cancer_brasil = diagnosticos_cancer_brasil.drop(' Total', axis=1)

#definir o index do dataframe
diagnosticos_cancer_brasil = diagnosticos_cancer_brasil.set_index('UF do tratamento') 

#dropar a linha Total 'Total' e 'Sem informação de Tratamento' por que não existem informações que ajudem na análise sobre os casos de Cancer de Mama / 2021
diagnosticos_cancer_brasil = diagnosticos_cancer_brasil.drop('Total', axis=0)
diagnosticos_cancer_brasil = diagnosticos_cancer_brasil.drop('Sem informação de tratamento', axis=0)
```

**Gráfico que exibe a quantidade de Tratamentos e a quantidade de Diagnósticos, para o Cancêr de Mama, realizados no estado de SP, no período de 2013 até 2021.**

> A idéia principal é exibir que houve uma diminuição, considerável, na quantidade de diagnósticos a partir do ano de 2020, início da pandemia.
A diminuição de diagnósticos, pode ter ocorrido devido ao fato da dificuldade de obter acesso à saúde ou mesmo, medo de procurar os hospitais.
A curva começa em declínio no ano de 2020, ano de início da Pandemia. 
Para efeito de comparação , é exibido o gráfico de quantidade de tratamentos , por Ano, para o cancêr de Mama, no estado de São Paulo.
O fato da curva está em declínio devido à pandemia é uma suposição, outros fatores desconhecidos podem ter interferido.

```python
#obter os casos de diagnósticos de cancer de Mama para o Estado de SP
diagnosticos_sp = diagnosticos_cancer_brasil[dados_tratamento_arquivo_brasil.index.str.contains('São Paulo')]
dados_diagnosticos_sp = diagnosticos_sp.T
dados_tratamento_sp   = dados_tratamento_arquivo_sp.T

fig = plt.figure(figsize=(15,5))
ax  =  fig.add_subplot(111)    # The big subplot
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(dados_diagnosticos_sp)
ax2.plot(dados_tratamento_sp)

fig.text(0.5, 0.0, 'Ano do Tratamento', ha='center', va='center')
fig.text(0.0, 0.5, 'Quantidade de Diagnósticos / Tratamentos', ha='center', va='center', rotation='vertical')

ax1.set_title('Casos de Cancer de Mama Diagnosticados no Estado de SP por Ano')
ax2.set_title('Casos de Cancer de Mama em Tratamento no Estado de SP por Ano')



plt.setp( ax.get_xticklabels(), visible=False)
plt.setp( ax.get_yticklabels(), visible=False)

plt.tight_layout()
```

**Gráfico que exibe a quantidade de Tratamentos e a quantidade de Diagnósticos, para o Cancêr de Mama, realizados na região Sudeste do Brasil, no período de 2013 até 2021.**


> A idéia principal é exibir que houve uma diminuição, considerável, na quantidade de diagnósticos a partir do ano de 2020, início da pandemia. A diminuição de diagnósticos, pode ter ocorrido devido ao fato da dificuldade de obter acesso à saúde ou mesmo, medo de procurar os hospitais. A curva começa em declínio no ano de 2020, ano de início da Pandemia. Para efeito de comparação , é exibido o gráfico de quantidade de tratamentos , por Ano, para o cancêr de Mama, na região Sudeste. Nota-se que o padrão do declínio é o mesmo do que do estado de SP. 
O fato da curva está em declínio devido à pandemia é uma suposição, outros fatores desconhecidos podem ter interferido.

```python
#Comparar quantidade de diagnósticos com quantidade de tratamentos para a região sudeste
dados_diagnosticos_regiao_sudeste = diagnosticos_cancer_brasil[diagnosticos_cancer_brasil.index.str.startswith('3')].T
dados_tratamento_regiao_sudeste   = dados_tratamento_arquivo_brasil[dados_arquivo_brasil.index.str.startswith('3')].T

fig = plt.figure(figsize=(15,5))
ax  =  fig.add_subplot(111)    # The big subplot
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(dados_diagnosticos_regiao_sudeste)
ax2.plot(dados_tratamento_regiao_sudeste)

fig.text(0.5, 0.0, 'Ano do Tratamento', ha='center', va='center')
fig.text(0.0, 0.5, 'Quantidade de Diagnósticos / Tratamentos', ha='center', va='center', rotation='vertical')

ax1.set_title('Casos de Cancer de Mama Diagnosticados na Região Sudeste')
ax2.set_title('Casos de Cancer de Mama em Tratamento na Região Sudeste')

plt.setp( ax.get_xticklabels(), visible=False)
plt.setp( ax.get_yticklabels(), visible=False)

l1 = plt.arrow(1,1, 3,1, head_width=0.2, color='gray', length_includes_head=True)
l2 = plt.arrow(1,1, 1,3, head_width=0.2, color='r', length_includes_head=True)
l3 = plt.arrow(1,1, 4,4, head_width=0.2, color='purple', length_includes_head=True)
l4 = plt.arrow(1,1, 4,4, head_width=0.2, color='b', length_includes_head=True)

plt.legend([l1, l2, l3, l4], ['São Paulo','Minas Gerais','Rio de Janeiro', 'Espirito Santo'], loc='upper left', bbox_to_anchor=(0.97, 1))

plt.tight_layout()
```
